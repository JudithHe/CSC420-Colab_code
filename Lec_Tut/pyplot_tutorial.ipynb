```python
from matplotlib import pyplot as plt
import numpy as np
import numpy.random as rand
```

## Line plots

```python
# state the starting point and ending point of the lines
x = [3, 2]
y = [1, 2]
plt.title('Plotting these values')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.plot(x, y, label='x and y')
plt.plot(x, [3, 4], 'r--', label='other')
plt.legend()
plt.show()
```

## Scatter plots

```python
# Generate data
mu1, mu2 = 2, 5
std = 1
sample1 = rand.normal(mu1, std, 50)
print(sample1.shape) # a row vector length is 50
sample2 = rand.normal(mu2, std, 50)

# Plot centers
fig = plt.gcf() #gcf create a new figure
fig.gca().add_artist(plt.Circle((mu1,mu1), radius=std, color='b', fill=False)) #gca create new axes for the figure
fig.gca().add_artist(plt.Circle((mu2,mu2), radius=std, color='g', fill=False))

# Plot point scatter
plt.plot(sample1[:25], sample1[25:], 'b*') #*makes them become dots
plt.plot(sample2[:25], sample2[25:], 'gD')
```

## Subplots

```python
# Simple data to display in various forms
x = np.linspace(0, 2 * np.pi, 400) # create a vector
y = np.sin(x ** 2)# use np to accerlerate


# Four axes, returned as a 2-d array
# axarr is an array of Axes objects if more than one subplot was created.
fig, axarr = plt.subplots(2, 2)
#print(axarr)
axarr[0, 0].plot(x, y)
axarr[0, 0].set_title('Axis [0,0]')
axarr[0, 1].scatter(x, y)
axarr[0, 1].set_title('Axis [0,1]')
axarr[1, 0].plot(x, y ** 2)
axarr[1, 0].set_title('Axis [1,0]')
axarr[1, 1].scatter(x, y ** 2)
axarr[1, 1].set_title('Axis [1,1]')
# Fine-tune figure; hide x ticks for top plots and y ticks for right plots
plt.setp([a.get_xticklabels() for a in axarr[0, :]], visible=False)
plt.setp([a.get_yticklabels() for a in axarr[:, 1]], visible=False)
plt.show()
```
