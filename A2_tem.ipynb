```
def Find_seam_path(img, min_col, sum_energy):
  #Make the connected path with smallest sum of enery Black on orginal img
  rows, cols = img.shape[0:2]
  #make the points in the path black points (Back_Trace)
  #range(start, end(not included, step))
  for r in range(rows-1, -1, -1):    
    img[r,min_col] = [0,0,0]
    if r != 0:
      if min_col == 0:
        if sum_energy[r-1,min_col+1]< sum_energy[r-1,min_col]:
          min_col = min_col+1
        elif min_col == cols -1:
          if sum_energy[r-1,min_col-1]< sum_energy[r-1,min_col]:
            min_col = min_col-1
        else:
          if sum_energy[r-1,min_col-1]<sum_energy[r-1,min_col] and sum_energy[r-1,min_col-1] < sum_energy[r-1,min_col+1]:
            min_col = min_col -1
          elif sum_energy[r-1,min_col+1]<sum_energy[r-1,min_col] and sum_energy[r-1,min_col+1] < sum_energy[r-1,min_col-1]:
            min_col = min_col +1
  return img
```
